## Primeiro Scraping

## Usando `requests`

```python
import requests
pagina = requests.get("https://evaldowolkers.wordpress.com/")
print(pagina.text)
#print(pagina.content)
```

## BeautifulSoup4

```python
import requests
import bs4

response = requests.get("https://evaldowolkers.wordpress.com/")
soup = bs4.BeautifulSoup(response.text, "html.parser")

print(soup.find_all('title'))
```

## Usando RegEx


```python
import re
from urllib.request import urlopen


html = urlopen("http://www.pythonparatodos.com.br/contato")
texto = html.read()

padrao = r'[\w.-]+@[\w.-]+'
resultado = re.findall(padrao, str(texto))
print(resultado)

padrao = r'\(\d{2}\) \d{5}\-\d{4}'
resultado = re.findall(padrao, str(texto))
print(resultado)
```
