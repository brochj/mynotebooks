### Criando a estrutura
No Next.js as próprias pastas são as rotas exemplo
```
/pages/index.js       --> site.com
/pages/contato.js     --> site.com/contato
/pages/blog/index.js  --> site.com/blog
/pages/sobre/oscar.js --> site.com/sobre/oscar
```
Então na raiz do projeto, criar `pages/index.js`

```js
// pages/index.js
import { useState } from 'react';

function Home(){
	return (
		<div>
			<h1>Home</h1>
			<Contador />
		</div>
	)
}

function Contador(){
	const [contador, setContador] = useState(1);

	function adicionarContador(){
		setContador(contador + 1)
	}

	return (
		<div>
			<div>{ contador }</div>
			<button onClick={adicionarContador}>Adicionar</button>
		</div>
	)
}

export default Home
```




# Navegação entre páginas (tradicional)
Na forma tradicional a cada navegacao tem que recomecar tudo do zero

Criar um Link com a tag `<a href="/sobre">`
```js
// pages/index.js
function Home(){
	return (
		<div>
			<h1>Home</h1>
			<a href="/sobre">Acessar página sobre</a>
		</div>
	)
}

export default Home
```
Criar um arquivo `sobre.js` 
```js
// pages/sobre.js
function Sobre(){
	return (
		<div>
			<h1>Sobre</h1>
			<a href="/">Acessar página Home</a>
		</div>
	)
}

export default Sobre
```

# Navegação entre páginas usando Next.js

Criar um Link utilizando o componente do Next.js



```js
// pages/index.js
import Link from 'next/Link';

function Home(){
	return (
		<div>
			<h1>Home</h1>
            <Link href="/sobre">
                <a>Acessar página sobre</a>
            </Link>
		</div>
	)
}

export default Home
```
Criar um arquivo `sobre.js` 

```js
// pages/sobre.js
import Link from 'next/Link';

function Sobre(){
	return (
		<div>
			<h1>Sobre</h1>
            <Link href="/">
                <a>Acessar página Home</a>
            </Link>
		</div>
	)
}

export default Sobre
```
